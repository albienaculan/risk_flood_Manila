```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely import wkt
from shapely.geometry import Point, Polygon
import osmnx as ox
import pyproj
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
import networkx as nx
```

```python
fp = "gadm41_PHL_3.shp"

ph_gdf = gpd.read_file(fp)
```

```python
ph_gdf = gpd.GeoDataFrame(ph_gdf, crs='epsg:4326')
```

```python
columns_of_interest = ["COUNTRY", "NAME_1", "NAME_2", "NAME_3", "TYPE_3", "ENGTYPE_3", "geometry" ]

ph_gdf_0 = ph_gdf[columns_of_interest]
```

```python
ph_gdf_0.rename(columns={"COUNTRY": "Country", 
                         "NAME_1": "province", 
                         "NAME_2": "City", 
                         "NAME_3": "Barangay", 
                         "geometry":"Geometry" }, inplace=True)
```

```python
ph_gdf_0
```

```python
ph_gdf_0.set_geometry("Geometry", inplace=True)
```

```python
ncr_gdf = ph_gdf_0[ph_gdf_0['province'] == 'Metropolitan Manila']
ncr_gdf.plot(figsize=(15,8), facecolor='none', edgecolor='black')
plt.axis('off')
plt.show()
```

```python
mnl_gdf = ph_gdf_0[(ph_gdf_0['City'] == 'Manila')]
mnl_gdf.plot(figsize=(15,8), facecolor='none', edgecolor='black')
plt.axis('off')
plt.show()
```

```python
mnl_gdf.head()
```

```python
census = pd.read_csv("2015_Census_of_Population.xlsx - Census.csv")

census.head()
```

```python
manila = pd.merge(mnl_gdf, census, on=["Barangay"])
```

```python
manila.head()
```

```python
# # result = pd.DataFrame(poi_ncr.groupby('city')['amenity'].value_counts())
# # result.columns = ['count']
# # result = result.reset_index()
# # result = result[result['amenity'] == 'restaurant']
# # result

result = pd.DataFrame(manila.groupby('District')['Total Population'].sum())
result.reset_index()

result.rename(columns={"Total Population":"District Population"}, inplace=True)

manila_total_population = pd.merge(manila, result, on=["District"])

manila_total_population.head()
```

```python
# merging the result df to the shapefiles, we have

fig, ax = plt.subplots(figsize=(15, 8))
manila_total_population.plot(ax=ax, column='District Population', cmap='YlOrRd', legend=True, edgecolor='black')
ax.axis('off')
plt.show()
```
