### Classifying Precipitation; Violent Precipitation

```python
# transaction date
df['DATETIME'] = pd.to_datetime(df['DATE'])
df['YEAR'] = df['DATETIME'].dt.year.astype(str)
df['MONTH'] = df['DATETIME'].dt.month
df['DAY'] = df['DATETIME'].dt.day


# convert month to string
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
             7: 'Jul', 8: 'Aug', 9: 'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
df['MONTH_'] = df['MONTH'].map(month_map)
df
```

```python
# Classify precipitation into light, moderate, heavy, violent
def get_part_of_day(prcp):
    """Return the part of day given the hour of day"""
    if prcp < 2.5:
        return 'light'
    elif (prcp >= 2.5) & (prcp < 10):
        return 'moderate'
    elif (prcp == 10) & (prcp < 50):
        return 'heavy'
    else:
        return 'violent'
#Based on the following source: https://en.wikipedia.org/wiki/Rain#cite_note-UKint-109
df.loc[:, 'RAINFALL_INTENSITY'] = df['PRCP'].apply(get_part_of_day)
df = df.fillna(0)
df
```

```python
sns.countplot(y='RAINFALL_INTENSITY',hue='RAINFALL_INTENSITY',data=df, order = df['RAINFALL_INTENSITY'].value_counts().index)
plt.title('Incidents of Violent Rainfall per Month')
```

```python
rainfall_df = df[df['RAINFALL_INTENSITY']=='violent']
sns.countplot(y='MONTH_',hue='RAINFALL_INTENSITY',data=rainfall_df, order = rainfall_df['MONTH_'].value_counts().index)
plt.title('Incidents of Violent Rainfall per Month')
```

Violent rainfall occurs most frequently from June to October, so the early-warning flood system will be active especially on those months.

### Streaks of Violent Precipitation

```python
for col in df.columns:
    print(col)
```

```python
to_drop = ['STATION', 'NAME', 'LATITUDE', 'LONGITUDE', 'ELEVATION', 
            'PRCP', 'PRCP_ATTRIBUTES', 'TMAX','TMAX_ATTRIBUTES', 'TAVG', 'TAVG_ATTRIBUTES',
           'TMIN', 'TMIN_ATTRIBUTES', 'DATETIME', 'YEAR', 'MONTH', 'DAY', 'MONTH_']
cdf = df.drop(to_drop, axis=1)
cdf.head()
```

```python
cdf['START_OF_STREAK'] = cdf.RAINFALL_INTENSITY.ne(cdf['RAINFALL_INTENSITY'].shift())
cdf
```

```python
cdf['STREAK_ID'] = cdf['START_OF_STREAK'].cumsum()
cdf
```

```python
cdf['STREAK_COUNTER'] = cdf.groupby('STREAK_ID').cumcount() + 1
cdf
cdf = cdf[cdf['RAINFALL_INTENSITY']=='violent']
cdf.head(30)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.countplot(data = cdf, x = 'STREAK_COUNTER').set(title = 'Length of Consecutive Days with Violent Rain')
```

Days with violent rain are often short-lived, but there have been a sizeable amount of incidents where violent rains would last from 2 to 4 days. Hence, preparations and safety-protocols to address typhoons that may last for the same duration should be prioritized.
