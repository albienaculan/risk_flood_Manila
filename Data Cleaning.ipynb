```python
import pandas as pd
```

```python
df = pd.read_csv("Weather_MNL.csv")

df.head()
```

```python
# Replace NaN with 0

df['PRCP'] = df['PRCP'].fillna(float(0))

df.head()
```

```python
# Get Date and Precipetation column

columns = ['DATE', 'PRCP']

df_precipation = df[columns]

df_precipation.head()
```

```python
# Complete the dates and replace them with 0

df_precipation['DATE'] = pd.to_datetime(df_precipation['DATE'])
df_precipation.set_index('DATE', inplace = True)

# print(df_precipation['DATE'].max()) #2022-11-19 00:00:00

idx = pd.date_range('2000-01-01','2022-11-19') # Date range of the data

df_precipation = df_precipation.reindex(idx, fill_value=0) # completes the dates and adds 0 if there is a missing date

df_precipation['PRCP'].value_counts()
```

```python
df_precipation.to_csv("PRCP_Manila.csv")
```
